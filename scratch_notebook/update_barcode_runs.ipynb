```python

#need to format sample names
lib = 'A';
date_selection = '240331'
rescue_num='4'; 

#notes rescue 1 was done on 13-Jan-2024; Rescue 2 was done 18-Jan-2024. Rescue 3 was done 09-Mar-2024

#either add string name of antibody (eg. 'RVA122'), say 'no_antibody', or 'VSVG' for VSV-G selections
#Then add '-#' where # is the replicate number. Eg replicate 1 is '-1' and so on.

conditions = ['CR57-20-1',
              'CR57-100-1',
              'CR57-200-1',
              'CR57-400-1',
              'CR57-1000-1',
              'CR57-2000-1'];


#make sample names
sample_names = []

for i in conditions:
    sample = lib+'-'+date_selection+'-'+'rescue_'+rescue_num+'-'+i;
    print(sample)
    sample_names.append(sample)

print(sample_names)

```

```python
#copy-paste directories -- match order for names

directories=['/shared/ngs/illumina/aaditham/240405_VH01189_267_AAFK7WYM5/Unaligned/Project_aaditham/LibA_CR57_20_S7_R1_001.fastq.gz',
             '/shared/ngs/illumina/aaditham/240405_VH01189_267_AAFK7WYM5/Unaligned/Project_aaditham/LibA_CR57_100_S8_R1_001.fastq.gz',
             '/shared/ngs/illumina/aaditham/240405_VH01189_267_AAFK7WYM5/Unaligned/Project_aaditham/LibA_CR57_200_S9_R1_001.fastq.gz',
             '/shared/ngs/illumina/aaditham/240405_VH01189_267_AAFK7WYM5/Unaligned/Project_aaditham/LibA_CR57_400_S10_R1_001.fastq.gz',
             '/shared/ngs/illumina/aaditham/240405_VH01189_267_AAFK7WYM5/Unaligned/Project_aaditham/LibA_CR57_1000_S11_R1_001.fastq.gz',
             '/shared/ngs/illumina/aaditham/240405_VH01189_267_AAFK7WYM5/Unaligned/Project_aaditham/LibA_CR57_2000_S12_R1_001.fastq.gz'];

             
```

```python
library='A';
date_value = 20240331;

libraries = [];
date = [];

for i in range(0, len(directories)):
    libraries.append(library);
    date.append(date_value);
```

```python
print(libraries)
print(date)
```

```python
#update barcode_runs_new
barcode_runs_new = pd.DataFrame()

barcode_runs_new['sample']=sample_names;
barcode_runs_new['library']=libraries;
barcode_runs_new['date']= date;
barcode_runs_new['fastq_R1']=directories;
```

```python
#okay, so note that we should not overwrite the old barcode_runs file with just the new info. So we need to concatenate.

#save barcode_runs_current as an archive in case of a mistake
barcode_runs_current.to_csv('barcode_runs_current_240408_CR57-added.csv');


dfs= [barcode_runs_current, barcode_runs_new];

barcode_runs_final = pd.concat(dfs)
```
